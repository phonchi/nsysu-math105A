### (16) Consider a variant of the Fibonacci sequence, which is defined as follows: (6%)

$$\begin{cases}
  \displaystyle F(0)=1, F(1)=4  \\
  \displaystyle F(n) = F(n-1) + F(n-2) \quad \text{n > 1} \\
  \end{cases}$$

The following program segment is designed to calculate the nth term of the above modified Fibonacci sequence:

````python
def fib_r(n):
  if n == 0:
    return 1
  elif n == 1:
    return 4
  else:
    return fib_r(n-1) + fib_r(n-2)    
````

Modify the above program to calculate the sequence using loop by completing the following function. Note your code should pass the following test data.

````python
def fib_i(n):
  """
    Parameters
    ----------
    n: int
        The index of the target number in the sequence
    Returns
    -------
    result : int
        the nth term of the number in the sequence 
  """
  n0 = 1
  n1 = 4
  if n == 0:
    return 1
  elif n == 1:
    return 4
  for i in range(n-1):
    result = n1 + n0
    n1, n0 = result, n1
  return result
````

> Ans: *double click here to answer the question.*

A stack is a restricted linear list in which all additions and deletions are made at one end, the top. The stacks are known as a last in, first out (LIFO) data structure.

A queue, on the other hand, is a linear list in which data can only be inserted at one end, called the rear, and deleted from the other end, called the front. A queue is a first in, first out (FIFO)  structure.

Stack can be used when we want to pairing the data For instance, when we type an expression with a lot of parentheses, we often forget to pair the parentheses. The compiler can then uses a stack to check that all opening parentheses are paired with a closing parentheses.



### (18) The linked list in our slide is defined as follows: (6%)

````python
class LinkedList:
  def __init__(self, nodes=None):
    self.head = None
    if nodes:
      node = Node(data=nodes.pop(0)) #pop will pop out the first entry of the list
      self.head = node
      for elem in nodes:
        node.link = Node(data=elem)
        node = node.link
  def __repr__(self): #This is to beautify printing, don't worry about this
    node = self.head
    nodes = []
    while node is not None:
      nodes.append(str(node.data))
      node = node.link
    nodes.append("None")
    return " -> ".join(nodes)

class Node:
  def __init__(self, data):
    self.data = data
    self.link = None
  def __repr__(self):
    return str(self.data)
````

Complete the following function to calculate the summation of two liked lists with the same length elementwise and store the results into a list. Note your code should pass the following test data.

````python
def SumLinkedList(ilist1, ilist2):
  """
    Parameters
    ----------
    ilist1: linked list
        The first input linked list that contains positive numbers.
    ilist2: linked list
        The second input linked list that contains positive numbers.
    Returns
    -------
    olist : list
        The output list that contains the summation of the two input linked list, note that summation is performed elementwise
  """
  # Your code here
  olist = []
  walker1 = ilist1.head
  walker2 = ilist2.head
  while walker1 != None:
    olist.append(walker1.data + walker2.data)
    walker1 = walker1.link
    walker2 = walker2.link  
  return olist
````

````python
# Test data
import numpy as np

ilist1 = LinkedList([3,5,7,9])
ilist2 = LinkedList([2,4,6,8])
ilist3 = LinkedList([3,11,12,15,19])
ilist4 = LinkedList([1,2,24,36,48])
assert(SumLinkedList(ilist1, ilist2)) == list(np.array([3,5,7,9])+np.array([2,4,6,8]))
assert(SumLinkedList(ilist3, ilist4)) == list(np.array([3,11,12,15,19])+np.array([1,2,24,36,48]))
````

Hint: The string can be sorted as an integer using `<` or `>` operator since Python will compare their ASCII code! You can access the score/name using `grades[idx][score]` or `grades[idx][name]`. In addition, **you can try to first sort the list according to key 1 then sort the resulting list according to key 2**.

````python
def bubble_sort(arr, keys):
  """
    Parameters
    ----------
    arr: list of dictionaries
        The unsorted input list of dictionaries
    keys: list of str
        Sort the data according to the key
    Returns
    -------
    arr : list of dictionaries
        The sorted list of dictionaries
  """
  n = len(arr)
  ## Move the wall one element to the right
  for key in keys[-1::-1]:
    for wall in range(n-1):
      ## bubble each element up to the right
      for j in range(n-1, wall, -1):
        if arr[j][key] < arr[j-1][key]:
          arr[j], arr[j-1] = arr[j-1], arr[j]
  return arr
````

### (25) The binary search tree in our slide is defined as follows: (6%)

````python
class tree:
  def __init__(self):
    self.data = 0
    self.left = None
    self.right = None
  def __repr__(self):
    lines, *_ = self._display_aux()
    return "\n".join(lines)
  def _display_aux(self):
    # https://stackoverflow.com/questions/34012886/print-binary-tree-level-by-level-in-python
    """Returns list of strings, width, height, and horizontal coordinate of the root."""
    # No child.
    if self.right is None and self.left is None:
      line = '%s' % self.data
      width = len(line)
      height = 1
      middle = width // 2
      return [line], width, height, middle

    # Only left child.
    if self.right is None:
      lines, n, p, x = self.left._display_aux()
      s = '%s' % self.data
      u = len(s)
      first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s
      second_line = x * ' ' + '/' + (n - x - 1 + u) * ' '
      shifted_lines = [line + u * ' ' for line in lines]
      return [first_line, second_line] + shifted_lines, n + u, p + 2, n + u // 2

    # Only right child.
    if self.left is None:
      lines, n, p, x = self.right._display_aux()
      s = '%s' % self.data
      u = len(s)
      first_line = s + x * '_' + (n - x) * ' '
      second_line = (u + x) * ' ' + '\\' + (n - x - 1) * ' '
      shifted_lines = [u * ' ' + line for line in lines]
      return [first_line, second_line] + shifted_lines, n + u, p + 2, u // 2

    # Two children.
    left, n, p, x = self.left._display_aux()
    right, m, q, y = self.right._display_aux()
    s = '%s' % self.data
    u = len(s)
    first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s + y * '_' + (m - y) * ' '
    second_line = x * ' ' + '/' + (n - x - 1 + u + y) * ' ' + '\\' + (m - y - 1) * ' '
    if p < q:
      left += [n * ' '] * (q - p)
    elif q < p:
      right += [m * ' '] * (p - q)
    zipped_lines = zip(left, right)
    lines = [first_line, second_line] + [a + u * ' ' + b for a, b in zipped_lines]
    return lines, n + m + u, max(p, q) + 2, n + u // 2    

def create_tree(root,val):  # create a binary tree
  newnode = tree()
  newnode.data = val
  newnode.left = None
  newnode.right = None
  if root == None:
    root = newnode
    return root
  else:
    current = root
    while current != None:
      backup = current
      if current.data > val:
        current = current.left
      else:
        current = current.right
    if backup.data > val:
      backup.left = newnode
    else:
      backup.right = newnode
  return root

data=[17,6,3,14,19,5,11,22,18,30]
ptr = None
root = None
for i in range(len(data)):
  ptr=create_tree(ptr,data[i])

ptr
````

Complete the following function so that it will return the nodes it has been visited when it match the target or the algorithm stops. Note that the nodes should be store into a list and your code should pass the following test data.

````python
def search(ptr,target):  
  """
    Parameters
    ----------
    ptr: tree
        The input binary search tree
    target: int 
        The target value.
    Returns
    -------
    L: list
        The nodes that has been visited
  """   
  L = []
  while True:
    if ptr == None:
      return L
    if ptr.data == target:
      L.append(ptr.data)
      return L
    elif ptr.data > target:
      L.append(ptr.data)
      ptr = ptr.left
    else:
      L.append(ptr.data)
      ptr = ptr.right
````

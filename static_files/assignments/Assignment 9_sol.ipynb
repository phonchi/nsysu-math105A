### (6)  Try to write the code that finds the greatest common divisor of `x` and `y` using a recursive algorithm where the recursive definition is as follows (The expression `x mod y` means dividing `x` by `y` and using the remainder as the result of the operation):
$$gcd(x,y)=\begin{cases}
  \displaystyle x  \quad \quad \quad \quad \quad \quad \quad \text{if y=0}\\
  \displaystyle gcd(y, x \bmod y) \quad \text{otherwise}
  \end{cases}$$



````python
def gcd(x,y):
  """
    Parameters
    ----------
    x: int
        The positive input integer
    y: int 
        The positive input integer
    Returns
    -------
    results : int
        The greatest common divisor of x and y
  """
  if y==0: 
    return x
  else:
    return gcd(y, x%y)
````

````python
import math
assert gcd(7,41) == math.gcd(7,41)
assert gcd(12,100) == math.gcd(12,100)
assert gcd(46,12) == math.gcd(46,12)
````

### (7) What are the major differences between an array and a linked list?

> Ans: *double click here to answer the question.*

- Array elements store in a contiguous memory location. Linked list elements can be stored anywhere in the memory or randomly stored. 
- Arrays allow random access and require less memory per element (do not need space for pointers) while lacking efficiency for insertion/deletion operations and memory allocation
- Linked list enjoys faster insertion/deletion Time while it has slower search times than arrays as random access is not allowed

### (9) Write the code that reverse the elements of an 1D array so that the last element becomes the first, the scond to the last becomes the second and so on. Do not use the built-in function like `reverse()`/`reversed()` for list or `flip()` for numpy array.



````python
# coding your answer here.
````

````python
def reverse(ilst):
  """
    Parameters
    ----------
    ilst: list
        The input list
    Returns
    -------
    olst : list
        The reversed list
  """
  i = 0
  j = len(ilst)-1
  while(i<j):
    ilst[i], ilst[j] = ilst[j], ilst[i]
    i = i+1
    j = j-1
  olst = ilst
  return olst
````

````python
assert(reverse([89,78,61,53,23,21,17,12,9,7,6,2,1])==list(reversed([89,78,61,53,23,21,17,12,9,7,6,2,1])))
assert(reverse([1,5,8,9])==list(reversed([1,5,8,9])))
assert(reverse([78,12,15,8,61,53,23,27])==list(reversed([78,12,15,8,61,53,23,27])))
assert(reverse([5])==list(reversed([5])))
````

### (10) Write the code that find the average of the numbers in a linked list of numbers.

````python
class LinkedList:
  def __init__(self, nodes=None):
    self.head = None
    if nodes:
      node = Node(data=nodes.pop(0)) #pop will pop out the first entry of the list
      self.head = node
      for elem in nodes:
        node.link = Node(data=elem)
        node = node.link
  def __repr__(self): #This is to beautify printing, don't worry about this
    node = self.head
    nodes = []
    while node is not None:
      nodes.append(str(node.data))
      node = node.link
    nodes.append("None")
    return " -> ".join(nodes)

class Node:
  def __init__(self, data):
    self.data = data
    self.link = None
  def __repr__(self):
    return str(self.data)
````

````python
def LinkedListAverage(ilist):
  """
    Parameters
    ----------
    ilist: llist
        The input list that contains numbers.
    Returns
    -------
    ave : int
        The average number
  """
  # Your code here
  walker = ilist.head
  counter = 0
  sum = 0
  while walker != None:
    sum = sum + walker.data
    walker = walker.link
    counter = counter + 1
  ave = sum/counter
  return ave
````

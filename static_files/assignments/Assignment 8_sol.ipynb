### (8) Write a simple program to find all the factors of a postive integer and return it as a list.

````python
# coding your answer here.
````

````python
def find_factors(number):
  """
    Parameters
    ----------
    number: int
        The positive input integer
    Returns
    -------
    factors : list
        The list that store all the factors
  """
  factors = []
  for i in range(1,number+1):
    if (number%i==0):
      factors.append(i)
  return factors
````

````python
assert(set(find_factors(12))==set([1,2,3,4,6,12]))
assert(set(find_factors(18))==set([1,2,3,6,9,18]))
assert(set(find_factors(3))==set([1,3]))
````

```python
def bubble_sort(arr):
  """
    Parameters
    ----------
    arr: list
        The unsorted input list that contains integers
    Returns
    -------
    arr : list
        The sorted list
  """
  n = len(arr)
  ## Move the wall one element to the right
  for wall in range(n-1):
    ## bubble each element up to the right
    for j in range(n-1, wall, -1):
      if arr[j] < arr[j-1]:
        arr[j], arr[j-1] = arr[j-1], arr[j]
  return arr
```

````python
def bubble_sort_r(arr):
  """
    Parameters
    ----------
    arr: list
        The unsorted input list that contains integers
    Returns
    -------
    arr : list
        The sorted list
  """
  n = len(arr)
  ## Move the wall one element to the right
  for wall in range(n-1):
    ## bubble each element up to the right
    for j in range(n-1-wall):
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]
  return arr
````

````python
assert(bubble_sort_r([89,78,61,53,23,21,17,12,9,7,6,2,1])==sorted([89,78,61,53,23,21,17,12,9,7,6,2,1]))
assert(bubble_sort_r([1,5,8,9])==sorted([1,5,8,9]))
assert(bubble_sort_r([78,12,15,8,61,53,23,27])==sorted([78,12,15,8,61,53,23,27]))
assert(bubble_sort_r([5])==sorted([5]))
````

### (10) Modify the selection sort program so that it can sort following list of scores.

```python
grades = [
        { 'name': 'david', 'score': 80},
```

Hint: The string can be sorted as integer using `<` or `>` operator since Python will compare their ASCII code! You can access the score/name using `grades[idx][score]` or `grades[idx][name]`

````python
grades = [
        { 'name': 'david', 'score': 80},
        { 'name': 'allen', 'score': 90},
        { 'name': 'bob',  'score': 100},
        { 'name': 'cathy', 'score': 85}
      ]
````

````python
def selection_sort(arr, key=None):
  """
    Parameters
    ----------
    arr: list of dictionaries
        The unsorted input list of dictionaries
    key: str
        Sort the data according to this key 
    Returns
    -------
    arr : list of dictionaries
        The sorted list of dictionaries
  """
  # Your code here
  n = len(arr)

  for wall in range(n-1):
    smallest = wall
    for cur in range(smallest+1,n):
      if arr[cur][key] < arr[smallest][key]:
        smallest = cur
    ## Swap smallest element with first element in the unsorted list
    arr[wall], arr[smallest] = arr[smallest], arr[wall]
  return arr
````
